```python
#Q.2
Scatter plot:A scatter plot is a set of points plotted on a horizontal and vertical axes. Scatter plots are important in statistics because
they can show the extent of correlation,if any, between the values of observed quantities .
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.scatter(x,y)
plt.show()
```

```python
#Q.3
The subplot() function takes three arguments thatdescribes the layout of the figure. The layout is organized in rows and columns, which are 
represented by the first and second argument.The third argument represents the index of the current plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,3,1)
plt.plot(x,y)
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(5,4,3)
plt.plot(x,y)
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(1,3,1)
plt.plot(x,y)
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(3,2,1)
plt.plot(x,y)
```

```python
#Q.4
Bar plot:Bar plot or charts are one of the many techniques used to present data in a visual form so that the reader may readily recognize patterns or trends. 
Bar plot or charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.show()
```

```python
plt.barh(company,profit)
plt.show()
```
